```python
def extract_text_from_pdf(pdf_path):
    """Extract text from a single PDF using pdfplumber"""
    print(f"Extracting text from: {pdf_path}")
    text = ""
    
    try:
        with pdfplumber.open(pdf_path) as pdf:
            for page_num, page in enumerate(pdf.pages):
                page_text = page.extract_text()
                if page_text:
                    text += f"--- Page {page_num + 1} ---\n{page_text}\n\n"
                else:
                    print(f"  Warning: No text found on page {page_num + 1}")
        
        return text
    
    except Exception as e:
        print(f"Error extracting {pdf_path}: {e}")
        return ""
```

```python
def clean_extracted_text(text, pdf_name):
    """
    Comprehensive text cleaning for journalism guides
    """
    print(f"Cleaning text from: {pdf_name}")
    
    # 1. Remove excessive whitespace and newlines
    text = re.sub(r'\n\s*\n', '\n\n', text)  # Multiple newlines to double newlines
    text = re.sub(r'[ \t]+', ' ', text)      # Multiple spaces to single space
    
    # 2. Fix common PDF extraction issues
    text = re.sub(r'(\w+)-\s*\n\s*(\w+)', r'\1\2', text)  # Fix hyphenated words
    text = re.sub(r'ﬁ', 'fi', text)  # Replace ligatures
    text = re.sub(r'ﬂ', 'fl', text)
    
    # 3. Remove page headers/footers (common patterns)
    text = re.sub(r'--- Page \d+ ---', '', text)  # Remove our page markers if needed
    text = re.sub(r'\d+\s*\n', '', text)  # Remove standalone page numbers
    
    # 4. Fix quotation marks and apostrophes
    text = re.sub(r'[``]', '"', text)  # Fix opening quotes
    text = re.sub(r"''", '"', text)    # Fix closing quotes
    text = re.sub(r'´', "'", text)     # Fix apostrophes
    
    # 5. Remove isolated characters (common in PDF extraction)
    text = re.sub(r'\n\s*[a-zA-Z]\s*\n', '\n', text)
     # 6. Standardize section headers (you can customize this)
    text = re.sub(r'\n([A-Z][A-Z\s]{5,})\n', r'\n## \1\n', text)  # Mark headers
    
    return text.strip()
```

```python
def save_cleaned_text(text, output_path, pdf_name):
    """Save cleaned text to file"""
    # Save as .txt for readability
    txt_path = output_path / f"{pdf_name}_cleaned.txt"
    with open(txt_path, 'w', encoding='utf-8') as f:
        f.write(text)
    
    # Also save metadata as JSON for your chatbot
    metadata = {
        "filename": f"{pdf_name}_cleaned.txt",
        "original_pdf": f"{pdf_name}.pdf",
        "character_count": len(text),
        "word_count": len(text.split()),
        "cleaning_applied": [
            "whitespace_normalization",
            "hyphenation_fixes", 
            "ligature_replacement",
            "quote_standardization"
        ]
    }
    
    json_path = output_path / f"{pdf_name}_metadata.json"
    with open(json_path, 'w', encoding='utf-8') as f:
        json.dump(metadata, f, indent=2, ensure_ascii=False)
    
    print(f"  ✓ Saved: {txt_path}")
    print(f"  ✓ Saved: {json_path}")
    return txt_path, json_path

    
```

```python

def process_all_pdfs():
    """Main function to process all PDFs"""
    
    # Set base directory for PDFs (read-only)
    base_dir = Path("/kaggle/input/extarct-pdf-using-pdfloder")
  
    
    # Set output directory to a writable location
    output_dir = Path("/kaggle/working/extracted_text")
    output_dir.mkdir(exist_ok=True)
    
    # Process each PDF
    results = {}
    
    # Use os.listdir to get all PDF files in the directory
    for pdf_file in base_dir.glob("*.pdf"):
        if pdf_file.is_file():
            print(f"\n{'='*50}")
            print(f"PROCESSING: {pdf_file.name}")
            print(f"{'='*50}")
            
            # Extract text
            raw_text = extract_text_from_pdf(pdf_file)
            
            if raw_text:
                # Clean text
                cleaned_text = clean_extracted_text(raw_text, pdf_file.stem)
                
                # Save results
                txt_path, json_path = save_cleaned_text(
                    cleaned_text, output_dir, pdf_file.stem
                )
                
                # Store in results
                results[pdf_file.stem] = {
                    "raw_text_length": len(raw_text),
                    "cleaned_text_length": len(cleaned_text),
                    "txt_path": str(txt_path),
                    "json_path": str(json_path),
                    "reduction_percent": round((1 - len(cleaned_text)/len(raw_text)) * 100, 2)
                }
                
                # Preview first 300 characters
                preview = cleaned_text[:300] + "..." if len(cleaned_text) > 300 else cleaned_text
                print(f"  Preview: {preview}")
            else:
                print(f"  ✗ Failed to extract text from {pdf_file.name}")
    
    # Save summary report
    summary_path = output_dir / "processing_summary.json"
    with open(summary_path, 'w', encoding='utf-8') as f:
        json.dump(results, f, indent=2, ensure_ascii=False)
    
    print(f"\n{'='*50}")
    print("PROCESSING COMPLETE!")
    print(f"{'='*50}")
    
    for pdf_name, stats in results.items():
        print(f"{pdf_name}:")
        print(f"  Raw: {stats['raw_text_length']} chars")
        print(f"  Cleaned: {stats['cleaned_text_length']} chars")
        print(f"  Reduction: {stats['reduction_percent']}%")
        print(f"  Saved to: {stats['txt_path']}")
    
    return results

if __name__ == "__main__":
    process_all_pdfs()
```

```python
from pathlib import Path

def load_cleaned_text_files(extracted_dir):
    """Load all cleaned text files and metadata"""
    extracted_dir = Path(extracted_dir)  # Convert to Path
    documents = []
    metadata_list = []
    
    # Get all cleaned text files
    for txt_file in extracted_dir.glob("*_cleaned.txt"):
        # Correct JSON file path
        json_file = txt_file.with_name(txt_file.stem.replace("_cleaned", "_metadata") + ".json")

        # Load text
        with open(txt_file, 'r', encoding='utf-8') as f:
            text = f.read()
        
        # Load corresponding metadata
        with open(json_file, 'r', encoding='utf-8') as f:
            metadata = json.load(f)
        
        # Create Document object
        doc = Document(
            page_content=text,
            metadata={
                "source": metadata.get("original_pdf", ""),
                "filename": metadata.get("filename", ""),
                "character_count": metadata.get("character_count", 0),
                "word_count": metadata.get("word_count", 0)
            }
        )
        
        documents.append(doc)
        metadata_list.append(metadata)
    
    return documents, metadata_list
```

```python
# Load JSON articles
def load_json_articles(json_path):
    json_path = Path(json_path)
    documents = []
    try:
        with open(json_path, 'r', encoding='utf-8') as f:
            articles = json.load(f)
        for article in articles:
            doc = Document(
                page_content=article.get("content", ""),
                metadata={
                    "source": article.get("url", ""),
                    "title": article.get("title", ""),
                    "date": article.get("date", "No date found"),
                    "type": "web_article"
                }
            )
            documents.append(doc)
    except Exception as e:
        print(f"Failed to load JSON articles: {e}")
    return documents
```

```python
all_documents = documents + json_documents
print(f"Loaded {len(all_documents)} documents (PDFs: {len(documents)}, JSON: {len(json_documents)})")
```

```python

# Create text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len
)

# Split documents into chunks
chunks = text_splitter.split_documents(all_documents)
print(f"Created {len(chunks)} chunks from {len(all_documents)} documents")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

# Create embeddings
embeddings = HuggingFaceEmbeddings(
    model_name="all-MiniLM-L6-v2",
    model_kwargs={'device': 'cpu'}
)
```

```python


# Create directory for vector store
os.makedirs("/kaggle/working/chroma_db", exist_ok=True)

# Create vector store
vectorstore = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory="/kaggle/working/chroma_db"
)

print("Vector store created and saved to /kaggle/working/chroma_db")
```

```python
# ========== ADD OPTIMIZED RETRIEVER HERE ==========
optimized_retriever = vectorstore.as_retriever(
    search_type="mmr",  # Maximal Marginal Relevance
    search_kwargs={
        "k": 3,
        "fetch_k": 10,  # Get more initially, then diversify
        "lambda_mult": 0.7  # Balance between relevance and diversity
    }
)
print("Optimized MMR retriever created")
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
import torch

# Load tokenizer and model
tokenizer_mistral = AutoTokenizer.from_pretrained("mistralai/Mistral-7B-Instruct-v0.2")
mistral_model = AutoModelForCausalLM.from_pretrained(
    "mistralai/Mistral-7B-Instruct-v0.2",
    device_map="auto",           # auto GPU allocation
    torch_dtype=torch.float16    # use half precision to save memory
)

# Create text-generation pipeline
mistral_pipe = pipeline(
    "text-generation",
    model=mistral_model,
    tokenizer=tokenizer_mistral,
    max_new_tokens=150,      # STRICTLY limit output
    min_new_tokens=50,       # Ensure minimum length
    do_sample=True,
    temperature=0.3,
    repetition_penalty=1.2,  # Prevent repetition
    pad_token_id=tokenizer_mistral.eos_token_id
)
```

```python
llm = HuggingFacePipeline(pipeline=mistral_pipe)
```

```python
from langchain_core.prompts import PromptTemplate

def create_qa_chain(llm, retriever):
    """
    Creates a simple QA chain that mimics RetrievalQA functionality
    """
    def answer_question(question):
        try:
            # Try the modern method first
            docs = retriever.invoke(question)
        except:
            try:
                # Try the older method name
                docs = retriever.get_relevant_documents(question)
            except:
                # Last resort - check what methods are available
                print("Available methods in retriever:", [method for method in dir(retriever) if not method.startswith('_')])
                raise
        
        # Format context from documents
        context = "\n\n".join([doc.page_content for doc in docs])
        
        # Create the prompt
        prompt_template = """
You are an AI assistant helping journalists write articles. Use ONLY the following information from the journalism guides to answer the user's question. If you don't have enough information in the provided context, say so.

Context:
{context}

Question: {question}

Helpful Answer:
"""
        prompt = prompt_template.format(context=context, question=question)
        
        # Get the answer
        answer = llm.invoke(prompt)
        
        return {
            "result": answer,
            "source_documents": docs
        }
    
    return answer_question
```

```python
# ========== FINALLY, CREATE YOUR QA CHAIN WITH THE OPTIMIZED RETRIEVER ==========
qa_chain = create_qa_chain(llm, optimized_retriever)  # Use optimized_retriever instead of vectorstore.as_retriever()
print("QA chain created with optimized retriever")
```

```python
eval_questions = [
    "How to structure a news article?",
    "What are the main goals of the inverted pyramid style?",
    "What advice is given regarding the use of jargon in reporting?"
]
eval_answers = [
    "A news article is structured using the inverted pyramid model, placing the most crucial information (the lede) first, followed by supporting details in descending order of importance.",
    "The main goals of the inverted pyramid style are to prioritize the most important information, ensure the reader gets the essential facts immediately, and facilitate easy trimming of the article by editors from the bottom up.",
    "Journalists should avoid jargon and technical language, or if necessary, ensure it is clearly explained for a general audience."
]
```

```python
# --- SIMULATION (A ADAPTER À VOTRE VECTORESTORE) ---

# IDs pertinents pour notre jeu de données (Hypothèse)
ground_truth_doc_ids = {
    "How to structure a news article?": ["doc_3"],
    "What are the main goals of the inverted pyramid style?": ["doc_1"],
    "What advice is given regarding the use of jargon in reporting?": ["doc_2"]
}

# --- Fonction de Notation ---
def evaluate_retrieval_manually(qa_chain, eval_questions, ground_truth_doc_ids):
    """Evaluate Hit Rate and MRR based on document IDs"""
    total_hit = 0
    total_mrr = 0

    for question in eval_questions:
        print(f"\n🔹 Question: {question}")
        result = qa_chain(question)
        retrieved_docs = result.get('source_documents', [])

        if not retrieved_docs:
            print("⚠️ Aucun document récupéré.")
            continue

        # ✅ Vérification de la présence des doc_id
        retrieved_ids = [doc.metadata.get('doc_id') for doc in retrieved_docs if doc.metadata.get('doc_id')]
        print(f"📄 IDs récupérés: {retrieved_ids}")

        pertinent_ids = ground_truth_doc_ids.get(question, [])
        print(f"🎯 Ground truth: {pertinent_ids}")

        found_pertinent = False
        for rank, doc_id in enumerate(retrieved_ids):
            if doc_id in pertinent_ids:
                total_hit += 1
                if not found_pertinent:
                    total_mrr += 1 / (rank + 1)
                    found_pertinent = True
                    break

    num_questions = len(eval_questions)
    print("\n--- 📊 Résultats globaux ---")
    print(f"Hit Rate moyen : {total_hit / num_questions:.4f}")
    print(f"MRR moyen      : {total_mrr / num_questions:.4f}")

# ⚠️ Vous devez adapter la ligne où l'ID du document est extrait (doc.metadata.get('doc_id')).

```
